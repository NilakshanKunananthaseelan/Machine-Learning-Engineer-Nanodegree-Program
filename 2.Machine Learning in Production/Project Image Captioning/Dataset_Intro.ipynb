#### **Intorduction to COCO dataset**

In this noteboook we will analyse the  Microsoft COCO dataset used for Image Captioning

```python
import os
from pycocotools.coco import COCO
```

```python
#initialize COCO API for caption annotations
val_data = 'val2014'
val_annfile = 'annotations/captions_val2014.json'
coco_val = COCO(val_annfile)

img_ids = list(coco_val.anns.keys())
```

```python
#plot a sample image
```

```python
import numpy as np 
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline

caption_id = np.random.choice(img_ids)
image_id = coco_val.anns[caption_id]['image_id']
img = coco_val.loadImgs(image_id)[0]
img_url = img['coco_url']
```
